```python
import pandas as pd
from sklearn.model_selection import train_test_split
david = pd.read_csv('/content/davids.csv')
y = pd.DataFrame(david['Accident_severity'])
x = pd.DataFrame(david['Number_of_casualties'])
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
y_train = y_train.astype(float)
regressor.fit(X_train, y_train)
```

```python
print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
#SAVING
import joblib
import pickle
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
joblib.dump(model,'linear_regression_model.pkl')
with open('linear_regression_model.pkl', 'wb') as file:
 pickle.dump(model, file)
```

```python
#TESTING
import joblib
import numpy as np
import pandas as pd
new_X = np.array([6, 7, 8, 9, 10]).reshape(-1, 1)
model = joblib.load('linear_regression_model.pkl')
predictions = model.predict(new_X)
new_data = pd.DataFrame({'X': new_X.flatten(), 'Predicted_Y': predictions.flatten()})
print(new_data)
```
